# Import Necesarry Library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

---
# Load the data

```python
house_data = pd.read_csv("house-price-data/train.csv")
```

```python
house_data.head(n=10)
```

# Question no 1 : Which Zone Has the Most Property Count?

```python
# create the data frame for property count analysis
MSZoning_count = house_data.groupby(["MSZoning"]).count()
MSZoning_zone = pd.DataFrame({"Zone" : list(MSZoning_count.index),
                              "Count_Property" : list(MSZoning_count["Id"].values)})

# create object figure and axes
fig, ax = plt.subplots(figsize=(10,7))     # figsize(width, height)

# create barplot
sns.set_style("darkgrid")
ax = sns.barplot(data=MSZoning_zone, x="Zone", y="Count_Property", palette="muted",
                 order=MSZoning_zone.sort_values('Count_Property', ascending=False).Zone)
ax.set_title("RL has the highest amount of property", fontsize=15, y=1)
ax.set_ylim([0,1300])
ax.set_xlabel("Zone", fontsize=16)
ax.set_ylabel("Amount of Property", fontsize=16)
for i in ax.containers:
    ax.bar_label(i, fontsize=16)
    
# show the plot
plt.suptitle("Amount of Property in Every Zones", fontsize=25)
plt.show()
```

```python
# create object figure and axes
fig, ax = plt.subplots(figsize=(10,7))

# create the pie chart
plt.pie(data = MSZoning_zone, x="Count_Property", labels="Zone",
       autopct='%.2f%%', pctdistance=1.2, labeldistance=1.4, textprops={'fontsize':14},
      wedgeprops={'linewidth':3, 'edgecolor':'white'})
ax.set_title("RL has the highest percentage of property", fontsize=15)

# add circle to the center of pie chart
my_circle = plt.Circle( (0,0), 0.7, color='white')
p = plt.gcf()
p.gca().add_artist(my_circle)

# show the plot
plt.suptitle("Amount of Property in Every Zones", fontsize=25)
plt.show()
```

### INSIGHT 

From the bar bar and pie chart we know that **Residential Low Density** zone has the biggest number of property **1151 property**, this is count for about **78.84%** of all property in Zones. **Commercial** zone has the smallest amount of property **10 property**, this is just count for about **0.68%** of all property in Zones

# Question Number 2 : Find price range in which the zone has the largest amount of property !

```python
# create function for customizing xticklabels
def x_ticklabels_custom(x_value, spacing):
    x_ticklabels = []
    for i,j in enumerate(x_value):
        if i%spacing == 0:
            x_ticklabels.append(str(j))
        else:
            x_ticklabels.append("")
    return x_ticklabels
```

## House Price Visualization of Residential Low Density Zone

```python
# prepare the data
RL_data = house_data[house_data["MSZoning"]=="RL"]

# calculate the central tendency of the data
RL_sale_mode = RL_data["SalePrice"].mode()[0]
RL_sale_median = RL_data["SalePrice"].median()
RL_sale_mean = RL_data["SalePrice"].mean()

# create figure and axes object
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,10), constrained_layout=True, gridspec_kw={'height_ratios':[5,1]},
                      sharex=True)

# create the distribution plot
bin_width = 20000
sns.histplot(data=RL_data, x="SalePrice", kde=True, binwidth=bin_width, ax=ax[0])
ax[0].set_title(f"Histogram BinWidth = ${bin_width}", fontsize=16)
ax[0].axvline(x=RL_sale_mode, color="red", linestyle="--", label=f"mode : ${RL_sale_mode:.0f}")
ax[0].axvline(x=RL_sale_median, color="green", linestyle="--", label=f"median : ${RL_sale_median:.0f}")
ax[0].axvline(x=RL_sale_mean, color="navy", linestyle="--", label=f"mean : ${RL_sale_mean:.0f}")
ax[0].set_ylabel("Count", fontsize=14)
ax[0].legend()

# create the boxplot
sns.boxplot(data=RL_data, x="SalePrice", ax=ax[1])
ax[1].set_xlabel("Sale Price ($)", fontsize=14)
xaxis_value = range(RL_data["SalePrice"].min(),RL_data["SalePrice"].max()+bin_width,bin_width)
ax[1].set_xticks(xaxis_value)
ax[1].set_xticklabels(x_ticklabels_custom(xaxis_value,5))

# show the plot
plt.suptitle("House Price Distribution Of Residential Low Density Zone", fontsize=20)
plt.show()
```

### INSIGHT

From the **House Price Distribution of Residential Low Density Zone** we could infer that the price range of the property is from aroud **\\$39300 - \\$359300**, because the rest of the data is outlier. We can use the **Price Mode** to state that the house price in Residential Low Density zone is **\\$135000**

# Question Number 3 : How is the House Price distibution of every zone compare to each other ?


## House Price Distribution Comparation of Every Zone

```python
# create figure and axes object
fig, ax = plt.subplots(figsize=(10,7))

# create median series for sorting boxplot
median_data = {"RL":RL_sale_median, "RM":RM_sale_median, "C (all)":C_sale_median, "FV":FV_sale_median, "RH":RH_sale_median}
median_series = pd.Series(median_data)

# plot the boxplot
sns.boxplot(data=house_data, y="SalePrice", x="MSZoning",
           order=median_series.sort_values(ascending=False).index)
ax.set_title("RL Price Distribution More Spread Than Other Zone", fontsize=15, y=1.02)
ax.set_ylabel("Sale Price ($)", fontsize=15)
ax.set_xlabel("Zone", fontsize=15)
y_value = range(house_data["SalePrice"].min(), house_data["SalePrice"].max(),50000)
ax.set_yticks(y_value)

# show the plot
plt.suptitle("Price Distribution Comparison Between Zone", fontsize=25)
plt.show()
```

### INSIGHT

From the **Price Distribution Comparison Between Zone** plot we could infer that from respective median the **most expensive** house price is at **Floating Village Residential Zone**, and the **most cheap** house is at **Commercial Zone**. Also House price at **Residential Low Density Zone** is the **most variative** than the other zones
